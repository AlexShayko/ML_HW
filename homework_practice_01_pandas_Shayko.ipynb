#### 6. [0.5 балла] Найдите курсы по выбору, на которые записывались как студенты 18-го года набора, так и студенты 17-го года.

```python
df_17_group = data[data['17_group'] != 0]
df_18_group = data[data['18_group'] != 0]
com_f1 = df_17_group[df_17_group['fall_1'].isin(df_18_group['fall_1'])].groupby('fall_1')['id'].count()
com_f2 = df_17_group[df_17_group['fall_2'].isin(df_18_group['fall_2'])].groupby('fall_2')['id'].count()
com_f3 = df_17_group[df_17_group['fall_3'].isin(df_18_group['fall_3'])].groupby('fall_3')['id'].count()
com_sp1 = df_17_group[df_17_group['spring_1'].isin(df_18_group['spring_1'])].groupby('spring_1')['id'].count()
com_sp2 = df_17_group[df_17_group['spring_2'].isin(df_18_group['spring_2'])].groupby('spring_2')['id'].count()
com_sp3 = df_17_group[df_17_group['spring_3'].isin(df_18_group['spring_3'])].groupby('spring_3')['id'].count()
com_total = pd.concat([com_f1, com_f2, com_f3, com_sp1, com_sp2, com_sp2]).index.unique()
print(list(com_total)) #выводим те курсы, которые являются общими у всех студентов
```

Методом исключения найдите курсы, которые предлагались только студентам 18-го года и только студентам 17-го года.

```python
# здесь идея в том, чтобы сначала найти все курсы для студентов 17го года и студентов 18го года, а затем найти уникальные
com_total = pd.Series(list(com_total))
comf1_17 = data[data['17_group'] != 0].groupby('fall_1').count()
comf2_17 = data[data['17_group'] != 0].groupby('fall_2').count()
comf3_17 = data[data['17_group'] != 0].groupby('fall_3').count()
coms1_17 = data[data['17_group'] != 0].groupby('spring_1').count()
coms2_17 = data[data['17_group'] != 0].groupby('spring_2').count()
coms3_17 = data[data['17_group'] != 0].groupby('spring_3').count()
all_cours_17 = pd.Series(list(pd.concat([comf1_17, comf2_17, comf3_17, coms1_17, coms2_17, coms3_17]).index.unique()))
unique_cours_17 = all_cours_17[~all_cours_17.isin(com_total)]
comf1_18 = data[data['18_group'] != 0].groupby('fall_1').count()
comf2_18 = data[data['18_group'] != 0].groupby('fall_2').count()
comf3_18 = data[data['18_group'] != 0].groupby('fall_3').count()
coms1_18 = data[data['18_group'] != 0].groupby('spring_1').count()
coms2_18 = data[data['18_group'] != 0].groupby('spring_2').count()
coms3_18 = data[data['18_group'] != 0].groupby('spring_3').count()
all_cours_18 = pd.Series(list(pd.concat([comf1_18, comf2_18, comf3_18, coms1_18, coms2_18, coms3_18]).index.unique()))
unique_cours_18 = all_cours_18[~all_cours_18.isin(com_total)]
print(list(unique_cours_17))
print(list(unique_cours_18))
```

### Визуализации и matplotlib

- plot — обычный поточечный график, которым можно изображать кривые или отдельные точки;
- hist — гистограмма, показывающая распределение некоторой величины;
- scatter — график, показывающий взаимосвязь двух величин;

В этом задании вы попробуете построить один из них. Не забывайте про базовые принципы построения приличных графиков:
- оси должны быть подписаны, причём не слишком мелко;
- у графика должно быть название;
